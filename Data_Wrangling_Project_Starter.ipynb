```python
import requests
import zipfile
import io
import os
import shutil
import tempfile

url = "https://files.grouplens.org/datasets/movielens/ml-latest-small.zip"
destination_dir = "datasets"

if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

response = requests.get(url)
zip_data = response.content

zip_stream = io.BytesIO(zip_data)

# Create a unique temporary directory
temp_dir = tempfile.mkdtemp()

with zipfile.ZipFile(zip_stream, "r") as zip_ref:
    zip_ref.extractall(temp_dir)

zip_stream.close()

# Remove the existing "datasets" directory (if it exists)
if os.path.exists(destination_dir) and os.path.isdir(destination_dir):
    shutil.rmtree(destination_dir)

# Move the temporary directory to "datasets"
shutil.move(temp_dir, destination_dir)
```

```python
import os

def list_directory_tree(root_dir, indent=""):
    for item in os.listdir(root_dir):
        item_path = os.path.join(root_dir, item)
        if os.path.isfile(item_path):
            print(indent + f"File: {item}")
        elif os.path.isdir(item_path):
            print(indent + f"Directory: {item}")
            list_directory_tree(item_path, indent + "  ")

root_directory = "datasets"

if os.path.exists(root_directory) and os.path.isdir(root_directory):
    print(f"Directory: {root_directory}")
    list_directory_tree(root_directory)
else:
    print(f"Directory {root_directory} does not exist.")
```

Visualization 1: Line Plot - Average Movie Ratings Over the Years

The line plot shows the trend of average movie ratings from the MovieLens dataset over the years of release. It helps us understand whether there is any correlation between the release year of movies and their ratings.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned datasets
netflix_data = pd.read_csv('datasets/cleaned_netflix_data.csv')
movie_lens_data = pd.read_csv('datasets/cleaned_movie_ratings.csv')

# Merge the datasets based on the movie title
merged_data = pd.merge(netflix_data, movie_lens_data, on='title', how='inner')

# Group the data by release year and calculate the average rating
average_ratings_by_year = merged_data.groupby('release_year')['rating'].mean().reset_index()

# Create a line plot to visualize the trend
plt.figure(figsize=(10, 6))
plt.plot(average_ratings_by_year['release_year'], average_ratings_by_year['rating'], marker='o')
plt.title('Average Movie Ratings Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

Visualization 2: Bar Chart - Netflix Content by Release Year

The bar chart displays the number of Netflix titles released in each year. It provides insights into how Netflix's content library has grown over time. This information is essential for understanding if there is a correlation between newer content and higher ratings.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned Netflix dataset
netflix_data = pd.read_csv('datasets/cleaned_netflix_data.csv')

# Group the data by release year and count the number of titles
count_by_year = netflix_data['release_year'].value_counts().reset_index()
count_by_year.columns = ['Release Year', 'Number of Titles']

# Create a bar chart to visualize the distribution
plt.figure(figsize=(10, 6))
plt.bar(count_by_year['Release Year'], count_by_year['Number of Titles'], color='skyblue')
plt.title('Netflix Content by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(axis='y')
plt.show()
```

Visualization 1: Stacked Bar Chart - Netflix Content Distribution by Genre Over the Years

The stacked bar chart displays the distribution of Netflix content by genre, grouped by release year. Each bar represents a specific year, and the height of each segment in the bar corresponds to the number of titles in that genre for that year. This visualization helps us understand how the distribution of content genres has evolved over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned Netflix dataset
netflix_data = pd.read_csv('datasets/cleaned_netflix_data.csv')

# Split the 'listed_in' column into individual genres
netflix_data['listed_in'] = netflix_data['listed_in'].str.split(', ')
netflix_data = netflix_data.explode('listed_in')

# Create a stacked bar chart to visualize the distribution by genre over the years
genre_distribution = netflix_data.groupby(['release_year', 'listed_in'])['show_id'].count().unstack()
genre_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Netflix Content Distribution by Genre Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Visualization 2: Pie Chart - Genre Distribution for the Most Recent Year

The pie chart shows the genre distribution of Netflix content for the most recent year in the dataset. It provides a snapshot of the content composition for the latest available year and can help identify the dominant genres.

```python
# Select the most recent year
latest_year = netflix_data['release_year'].max()

# Filter data for the latest year
latest_year_data = netflix_data[netflix_data['release_year'] == latest_year]

# Count the number of titles in each genre
genre_counts = latest_year_data['listed_in'].value_counts()

# Create a pie chart to visualize the genre distribution for the latest year
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Genre Distribution on Netflix for {latest_year}')
plt.show()
```
